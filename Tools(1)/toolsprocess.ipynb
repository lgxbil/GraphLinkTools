```python
import sys
import pickle
import multiprocessing
import json
import os
from tqdm import tqdm
import re
import glob
from transformers import AutoTokenizer

def count_tokens(text):
    tokenizer = AutoTokenizer.from_pretrained("/home/snrobot/lin/Llm-model/qwen2.5-7b")
    # 根据 text 的类型进行处理
    if isinstance(text, (dict, list)):
        # 如果是字典或列表，转换为 JSON 字符串
        text_str = json.dumps(text, ensure_ascii=False, indent=4)
    elif isinstance(text, str):
        # 如果是字符串，直接使用
        text_str = text
    else:
        # 其他类型（如数字、布尔值等），转换为字符串
        text_str = str(text)
    
    # 使用 tokenizer 编码并计算 Token 数量
    tokens = tokenizer.encode(text_str)
    token_count = len(tokens)
    return token_count
```

```python
import json
import os
import glob
from typing import List
api_folder_path = "/home/snrobot/lin/GraphLinkTools/Tools(1)/func_response"
info_files = glob.glob(os.path.join(api_folder_path, '*.json'))
for file in info_files:
    with open(file, 'r') as file:
        print(f"正在处理文件: {file}")
        data = json.load(file)
        res = data["row"]
        if isinstance(res["required"], List):
            cnt = count_tokens(res["required"])
            print(f"文件: {file}, Token 数量: {cnt}")
            continue
        if res["required"].get("data", False) and "error" not in res["required"]["data"]:
            cnt = count_tokens(res["required"]["data"])
            print(f"文件: {file}, Token 数量: {cnt}")
            continue
        elif res["optional"].get("data", False) and "error" not in res["optional"]["data"]:
            cnt = count_tokens(res["optional"]["data"])
            print(f"文件: {file}, Token 数量: {cnt}")
```

```python
# 对api进行分类操作
import json
import os
import glob

api_folder_path = "/home/snrobot/lin/GraphLinkTools/Tools(1)/required_params_link(3)"
output_folder_path = "/home/snrobot/lin/GraphLinkTools/Tools(1)/classify_api_link"

# 确保输出目录存在
os.makedirs(output_folder_path, exist_ok=True)

# 定义需要分类的字段
categories = ["hotels", "meta", "car", "taxi", "flights", "attraction", "restaurants"]

# 为每个分类创建文件夹
for category in categories:
    os.makedirs(os.path.join(output_folder_path, category), exist_ok=True)

# 获取所有 JSON 文件
info_files = glob.glob(os.path.join(api_folder_path, '*.json'))

for file_path in info_files:
    with open(file_path, 'r') as file:
        print(f"正在处理文件: {file_path}")
        data = json.load(file)
        endpoint = data.get("docs_row", {}).get("endpoint", "")

        # 检查 endpoint 中是否包含特定字段
        for category in categories:
            if category in endpoint:
                output_path = os.path.join(output_folder_path, category, os.path.basename(file_path))
                with open(output_path, 'w') as output_file:
                    json.dump(data, output_file, indent=4)
                print(f"文件已保存到: {output_path}")
```
